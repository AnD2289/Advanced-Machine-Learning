## Data Structures in Python
Python comes equipped with some useful built-in data structures, broadly similar to Java's collections package.

## 1.) Lists: 

Lists store a sequence of mutable items, it's a modified array structure with added-functionalities.

We can use the + operator to do list concatenation.

Python also allows negative-indexing from the back of the list.

Lists can be sliced.

Nested lists can represent multi-dimensional data.

```python
a = ['orange','apple',1,3]
```

## 2.) Tuples:
A data structure similar to the list is the tuple, which is like a list except that it is immutable once it is created (i.e. you cannot change its content once created). Note that tuples are surrounded with parentheses while lists have square brackets.

## 3.) Dictionaries:

Dictionary stores a map from one type of object (the key) to another (the value). The key must be an immutable type (string, number, or tuple). The value can be any Python data type.

## 3.) Sets:

A set is another data structure that serves as an unordered list with no duplicate items. Below, we show how to create a set, add things to the set, test if an item is in the set, and perform common set operations (difference, intersection, union):

```python
shapes = ['circle','square','triangle','circle']
setOfShapes = set(shapes)
setOfShapes
```

```python
setOfShapes.add('polygon')
setOfShapes
```

```python
'circle' in setOfShapes
```

```python
'rhombus' in setOfShapes
```

```python
favoriteShapes = ['circle','triangle','hexagon']
setOfFavoriteShapes = set(favoriteShapes)
setOfShapes - setOfFavoriteShapes
```

```python
setOfShapes & setOfFavoriteShapes
```

```python
setOfShapes | setOfFavoriteShapes
```

## Control Flow & Logic Statements:

In Python, we have For and do-while loops, conditional statements, and functions.

```python
# This is what a comment looks like
fruits = ['apples','oranges','pears','bananas']
for fruit in fruits:
    print(fruit + ' for sale')

fruitPrices = {'apples': 2.00, 'oranges': 1.50, 'pears': 1.75}
for fruit, price in fruitPrices.items():
    if price < 2.00:
        print('%s cost %f a pound' % (fruit, price))
    else:
        print(fruit + ' are too expensive!')
```

For functional programming we may use lamda with map or filter:

```python
map(lambda x: x * x, [1,2,3])
```

```python
filter(lambda x: x > 3, [1,2,3,4,5,4,3,2,1])
```

# Section 2. Numpy & Matplotlib:

## Intro to Numpy

NumPy is a Python library used for working with arrays.

In Python we have lists that serve the purpose of arrays, but they are slow to process.

NumPy aims to provide an array object that is up to 50x faster than traditional Python lists.

The array object in NumPy is called ndarray, it provides a lot of supporting functions that make working with ndarray very easy.


```python
import numpy as np
```

```python
a = np.array([5, 6, 7])
a
```

```python
# We can get the dimensions of an array object as follows:

print(np.shape(a))
```

We can slice a numpy array just like a list

```python
c = a[-3:-1]
c
```

Copy method in numpy:

```python
a = np.array([1, 2, 3])
b = a
a[0] = -20
a
```

```python
b
```

```python
c = np.copy(a)
a[0] = 50
c
```

Array concatenation in numpy:

```python
arr1 = np.array([[1, 2], [3, 4]])

arr2 = np.array([[5, 6], [7, 8]])

arr = np.concatenate((arr1, arr2), axis=1)

print(arr)
```

## Intro to Matplotlib

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
```

You can use numpy alongside Maplotlib for graphical representation which is helpful in observing Training/ Testing errors

```python
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#To plot points without lines you can use shortcut string notation parameter 'o'
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
# You can plot as many lines as you like by simply adding more plt.plot() functions:

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

#labels can be added as:

plt.title("Your title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


plt.plot(x1, y1, x2, y2)
plt.show()
```

## Subplots

The subplots() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```
