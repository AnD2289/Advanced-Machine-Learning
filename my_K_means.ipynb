```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import numpy as np

X = np.array([[1, 2],
              [2, 2],
              [2, 1],
              [-1, 5],
              [-2, -1],
              [-1,-1]])

plt.scatter(X[:,0], X[:,1], s=150)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
colors = 10*["g","r","c","b","k"]
```

```python
X = np.array([[1, 2],
              [2, 2],
              [2, 1],
              [-2, -1],
              [-1,-1],
              [-1, 5]])

plt.scatter(X[0:5,0], X[0:5,1], s=100, color=['red'])
plt.scatter(0.4, 0.6, s=150, marker='x',color=['red'])
plt.scatter(X[5,0], X[5,1], s=100,color=['blue'])
plt.scatter(-1, 5, s=150, marker='x',color=['blue'])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```
